```python
def moyenne_positifs(nombres):
    positifs = [n for n in nombres if n > 0]
    if not positifs:

    return sum(positifs) / len(positifs)
```

```python
def nettoyer_revenus(bruts):
    resultat = []
    for r in bruts:
        try:
            v = int(r)
            if v < 0:
                resultat.append(None)
            else:
                resultat.append(v)
        except (ValueError, TypeError):
            resultat.append(None)
    return resultat
```
