## Exercice 2 — arithmétique

Calcule la TVA à 20 % sur un prix hors taxe de 129,99 €. Affecte le résultat à une variable `tva`.

````python
# TON CODE ICI

prix_ht = 129.99
tva = None  # calcule la TVA à 20 %
````

````python
# === Vérification ===

import math

assert tva is not None, "`tva` vaut None — tu dois calculer le résultat."
assert isinstance(tva, (int, float)), f"`tva` doit être un nombre, pas {type(tva).__name__}."
attendu = 25.998
assert math.isclose(tva, attendu, rel_tol=1e-6), (
    f"Attendu ~{attendu}, reçu {tva}. Rappel : TVA = prix_ht × taux, avec taux = 0.20."
)
print(f"✅ OK — TVA = {tva:.2f} €")
````

## Exercice 4 — modulo

````python
# TON CODE ICI

nombre = 42
est_pair = None
````
